```python
# Python code for the Sales Analysis using Linear Regression

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data (for demonstration purposes)
marketing_spend = np.array([1000, 2000, 3000, 4000, 5000]).reshape(-1, 1)
sales = np.array([100, 190, 280, 370, 460])

# Linear Regression Model
model = LinearRegression().fit(marketing_spend, sales)

# Predicted Sales
predicted_sales = model.predict(marketing_spend)

# Visualization
plt.scatter(marketing_spend, sales, color='blue', label='Actual Sales')
plt.plot(marketing_spend, predicted_sales, color='red', label='Predicted Sales')
plt.title('Canopy Growth Sales Analysis')
plt.xlabel('Marketing Spend in $USD')
plt.ylabel('Sales in $USD')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes
nodes = ["Product Demand", "Sales", "Marketing", "Product Supply", "Product Availability", 
         "Price", "Competitor Activity", "Market Trends", "Feedback from Customers", "Production Costs"]
G.add_nodes_from(nodes)

# Add edges
edges = [("Product Demand", "Sales"), ("Sales", "Marketing"), ("Marketing", "Product Demand"),
         ("Product Supply", "Product Availability"), ("Product Availability", "Price"),
         ("Price", "Product Demand"), ("Competitor Activity", "Product Demand"),
         ("Market Trends", "Product Demand"), ("Feedback from Customers", "Sales"),
         ("Production Costs", "Price")]
G.add_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", width=2, edge_color="gray")
plt.title("Causal Loop Diagram for Domestic Sales")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G2 = nx.DiGraph()

# Add nodes
nodes2 = ["Stock of Products", "Inflow", "Outflow", "Production Rate", "Sales Rate", 
          "Foreign Demand", "Export Costs", "Currency Exchange Rates", "Regulations and Tariffs", 
          "Competitor Activity in Foreign Markets", "Marketing in Foreign Markets", "Feedback from Foreign Customers", 
          "Profitability", "Product Modifications"]
G2.add_nodes_from(nodes2)

# Add edges
edges2 = [("Inflow", "Stock of Products"), ("Stock of Products", "Outflow"), ("Foreign Demand", "Outflow"),
          ("Export Costs", "Profitability"), ("Currency Exchange Rates", "Profitability"),
          ("Regulations and Tariffs", "Inflow"), ("Regulations and Tariffs", "Outflow"),
          ("Competitor Activity in Foreign Markets", "Foreign Demand"), ("Marketing in Foreign Markets", "Foreign Demand"),
          ("Feedback from Foreign Customers", "Product Modifications")]
G2.add_edges_from(edges2)

# Define the shell layout
shell_layout = [["Stock of Products", "Inflow", "Outflow", "Production Rate", "Sales Rate", "Profitability", "Product Modifications"],
                ["Foreign Demand", "Export Costs", "Currency Exchange Rates", "Regulations and Tariffs"],
                ["Competitor Activity in Foreign Markets", "Marketing in Foreign Markets", "Feedback from Foreign Customers"]]

# Draw the graph using the shell layout
plt.figure(figsize=(12, 8))
nx.draw_shell(G2, nlist=shell_layout, with_labels=True, node_size=3000, node_color="lightgreen", font_size=10, font_weight="bold", width=2, edge_color="gray")
plt.title("CGC Stock and Flow Model for Foreign Sales")
plt.show()
```
